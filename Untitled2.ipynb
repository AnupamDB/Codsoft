```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv("/content/creditcard.csv")
```

```python
#data.fillna(data.median(),inplace=True)
data=data.dropna(axis=1)
```

```python
x=data.drop(['Class'], axis=1)
y=data['Class'].values.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
# # Convert labels to integers if they are not already
# x_train = x_train.astype(int)
# x_test = x_test.astype(int)
```

```python
# Create a mask to filter values other than 0 and 1
mask = (y_train == 0) | (y_train == 1)
mask1 = (x_train == 0) | (x_train == 1)
# Apply the mask to filter the training data
x_train = x_train[mask1]
y_train = y_train[mask]

# Similarly, apply the mask to the test data
mask_test = (y_test == 0) | (y_test == 1)
mask_test1 = (x_test == 0) | (x_test == 1)
x_test = x_test[mask_test1]
y_test = y_test[mask_test]
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='auto', random_state=42)
x_resampled, y_resampled = smote.fit_resample(x, y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_resampled, y_resampled, test_size=0.2, random_state=42)
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train, y_train)
y_pred=model.predict(x_test)
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

```python
print("Accuracy Score: ", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_pred)
```
